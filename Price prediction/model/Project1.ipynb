```python
df1['BHK'] = df1['size'].apply(lambda x: int(x.split(' ')[0]))
df1
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df1[~df1['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df2 = df1.copy()
df2['total_sqft'] = df2['total_sqft'].apply(convert_sqft_to_num)
df2 = df2[df2.total_sqft.notnull()]
df2
```

```python
df3 = df2.copy()
df3['price_per_sqft'] = df3.price * 100000 / df3.total_sqft
df3.head()
```

```python
len(df3.location.unique())
```

```python
location_stats = df3.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats
```

```python
len(location_stats[location_stats <= 10])
```

```python
len(location_stats[location_stats > 10])
```

```python
location_stats_less_than_10 = location_stats[location_stats <= 10]
df3.location = df3.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df3.location.unique())
```

```python
df3.head()
```

```python
df3[df3.total_sqft / df3.BHK < 300]
```

```python
df4 = df3[df3.total_sqft / df3.BHK >= 300]
df4.shape
```

```python
df4.price_per_sqft.describe()
```

```python
df5 = df4[(df4.price_per_sqft < (df4.price_per_sqft.mean() + 1 * df4.price_per_sqft.std())) 
    & (df4.price_per_sqft > (df4.price_per_sqft.mean() - 1 * df4.price_per_sqft.std()))]
df5.shape
```

```python
df5 = df5[df5.bath <= df5.BHK + 2]
df5.shape
```

```python
import matplotlib
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.BHK==2)]
    bhk3 = df[(df.location==location) & (df.BHK==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
    
plot_scatter_chart(df5,"Rajaji Nagar")
```

```python
plot_scatter_chart(df5,"Hebbal")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df6 = remove_bhk_outliers(df5)
df6.shape
```

```python
plot_scatter_chart(df6,"Rajaji Nagar")
```

```python
plot_scatter_chart(df6,"Hebbal")
```

```python
df7 = df6.drop(['size', 'price_per_sqft'], axis = 'columns')
df7.head()
```

```python
df7 = pd.get_dummies(df7, drop_first = True)
df7
```

```python
X = df7.drop('price', axis = 'columns')
y = df7.price
X.head()
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
X_train, X_test, y_train, y_test = tts(X, y, train_size = 0.8)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
cross_val_score(LinearRegression(), X, y).mean()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
model_params = {
    'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
}
```

```python
scores = []
    #cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
for algo_name, config in model_params.items():
    gs =  GridSearchCV(config['model'], config['params'], return_train_score=False)
    gs.fit(X,y)
    scores.append({
        'model': algo_name,
        'best_score': gs.best_score_,
        'best_params': gs.best_params_
    })
    
results = pd.DataFrame(scores,columns=['model','best_score','best_params'])
results
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
```

```python
predict_price('Indira Nagar',1000, 2, 3)
```
